## beautiful soup 資料萃取

```python
from bs4 import BeautifulSoup
#import bs4

html_sample = '''
<html> 
  <body> 
    <h1 id="title">Hello World</h1> 
    <a href="#" class="link">This is link1</a> 
    <a href="# link2" class="link">This is link2</a> 
  </body>
</html>
'''
soup = BeautifulSoup(html_sample, 'html.parser')
print(type(html_sample))
print(type(soup))
```

```python
h1 = soup.select('h1')
print(h1)
print(type(h1))
```

```python
print(h1[0])
print(type(h1[0]))
print(h1[0].text)
```

```python
a = soup.select('a')
print(a[0].text)
print(a[1].text)
```

## 根據ＣＳＳ取值

```python
# <h1 id="title">Hello World</h1> 
print(soup.select('h1')) # 根據標籤為h1 
print(soup.select('#title')) # 根據id 為 title
title = soup.select('#title')
print(title[0].text)
```

```python
#  <a href="#" class="link">This is link1</a> 
#  <a href="# link2" class="link">This is link2</a> 

print(soup.select('a'))# 根據標籤為a
print(soup.select('.link'))# 根據css 為 link
```

```python
links = soup.select('.link')
print(links[0])
print(links[1])
print(links[0].text)
print(links[1].text)
```

```python
for li in links:
    print(li) ## 程式縮排即迴圈內 四個空白
```

```python
for li in links:
    print(li.text) ## 裡面的內容爬出來
```
